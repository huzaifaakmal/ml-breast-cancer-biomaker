# Step 0: Project Overview & Executive Summary

**Author:** Huzaifa Akmal (Sanata Dharma University)
**Supervisor:** Narendar Kumar (University of Technology of Belfort-Montbéliard, UTBM)

## 1. Research Goal
This study aims to identify robust gene expression signatures for breast cancer and develop a machine learning classifier capable of distinguishing tumor from normal tissue across different experimental platforms.

## 2. Technical Challenge: The "Batch Effect"
Biological data from different labs often contains technical noise (batch effects). This project implements **Independent Z-score Normalization** and a **Robust Scikit-Learn Pipeline** to ensure the model learns biology, not lab-specific noise.

## 3. Key Findings
- **Final Model:** Balanced Linear SVM (5-gene Signature).
- **Performance:** ROC-AUC of **0.89** on external validation.
- **Top Biomarkers:** FHL1, PCK1, ACVR1C, FABP4, and MT1H.
