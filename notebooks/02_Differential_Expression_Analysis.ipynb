# Volcano Plot Visualization
This notebook generates a volcano plot to visualize the results of Differential Expression Analysis (DEA). 

**Purpose:** A volcano plot helps identify genes that are both highly significant (low P-value) and have a large magnitude of change (high Fold Change).
- **Red dots:** Significantly Upregulated genes.
- **Blue dots:** Significantly Downregulated genes.
- **Gray dots:** Not significant genes.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1. Load DEG Results
df = pd.read_csv("../raw_data/GSE42568_DEG_results.tsv", sep="\t")

# 2. Calculate -log10 of adjusted P-values
df["neg_log10_adjP"] = -np.log10(df["adj.P.Val"])

# 3. Categorize genes based on thresholds
fc_threshold = 1
p_threshold = 0.05

def categorize(row):
    if row["adj.P.Val"] < p_threshold and row["logFC"] > fc_threshold:
        return "Upregulated"
    elif row["adj.P.Val"] < p_threshold and row["logFC"] < -fc_threshold:
        return "Downregulated"
    else:
        return "Not Significant"

df["Category"] = df.apply(categorize, axis=1)

# 4. Plotting
plt.figure(figsize=(10, 8))

colors = {
    "Upregulated": "red",
    "Downregulated": "blue",
    "Not Significant": "gray"
}

for category in colors:
    subset = df[df["Category"] == category]
    plt.scatter(subset["logFC"], subset["neg_log10_adjP"], 
                c=colors[category], label=category, s=15, alpha=0.6)

# Add significance lines
plt.axvline(x=fc_threshold, linestyle="--", color='black', alpha=0.5)
plt.axvline(x=-fc_threshold, linestyle="--", color='black', alpha=0.5)
plt.axhline(y=-np.log10(p_threshold), linestyle="--", color='black', alpha=0.5)

plt.xlabel("log2 Fold Change")
plt.ylabel("-log10 Adjusted P-value")
plt.title("Volcano Plot: Breast Cancer (Tumor vs Normal)")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
